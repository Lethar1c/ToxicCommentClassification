## Модель 1. MLP

Попробуем собрать N самых часто используемых слов и на основании их сделаем вывод о токсичности того или иного комментария


### Преимущества
- нет

### Недостатки
- Много весов (ОЧЕНЬ МНОГО)
- Риск переобучения
- До жути малое количество данных
- Потеря контекста (например You're a devil и Devil is Bible character)

## 1. Обработка данных

Напишем класс, который при инициализации ищет топ N самых частых слов, затем для каждого комментария возвращает вектор длины N, где значение = 1, если слово есть, и 0 иначе
