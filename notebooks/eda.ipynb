```python
df['summarize'] = (df['toxic'] + df['severe_toxic'] + df['obscene'] + df['threat'] + df['insult'] + df['identity_hate'])
```

Примерно 11-12% комментариев - токсичны. При этом в среднем присутствует минимум две метки

```python
df[(df['summarize'] >= 1) & (df['toxic'] == 0)]
```

У нас присутствуют каждая категория по отдельности, т.е. метка toxic -- не панацея и не универсальная таблетка. Создадим колонку negative = summarize > 0 -- комментарии, в которых присутствует хоть что-то негативное. Остальные удалим в целях упрощения учебного проекта

```python
df = df.drop(axis='columns', labels=['id', 'toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'])
```

```python
df = df.copy()
df['negative'] = (df['summarize'] > 0).astype(int)
```

```python
df.describe()
```

```python
df = df.drop(axis='columns', labels=['summarize'])
```

```python
df.to_csv("./../data/processed/train.csv")
```
